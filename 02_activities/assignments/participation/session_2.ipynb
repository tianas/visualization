For each visualization, discuss the following questions:

- **Visualization 1: US Gun Killings in 2018**. This is an interactive plot that shows lifespans of those killed by gun violence in 2018 in the US. Each person has a trajectory that is coloured yellow up until their age of death, which turns to white to show potential lifespans (i.e., projected "stolen years"). From this visualization, we can learn that many gun deaths were concentrated within the adult years (i.e., few points end within the first fifth or so of the plot). When hovering over the x-axis, we can see density distributions of the two variables (i.e., density around the average age of death from gun violence vs. projected mortality rate from other causes). The data points below the plot on the webpage provide snapshots of the main points from the plot as well. This is not an objective, neutral data visualization. The "stolen years" variable is a projected calculation included to frame the data in a way that speaks to tragedy and loss (the value itself shows nearly half a million stolen years from these individuals because of gun violence).

- **Visualization 2: Washington Post Active Shooters**. This is a bar plot showing the number of active shooter incidents per year from 2000-2015 in the US. We can learn that there are more consistently high active shooter events in the second half of the plot, highlighting that active shooter incidences were rising. This is a more objective, neutral example, as it shows the data in a very simple manner without projection or valenced language. The simplicity allows the reader to glean the facts and draw their own conclusions.
